```python
# Load
pca   = joblib.load('pca.joblib')
vectorizer = joblib.load('tfidf.joblib')
Scaler = joblib.load('scaler.joblib')
```

```python
def everything():
    x_nontest = df.copy()
    
    #vectorize
    nontestIDF = vectorizer.transform(x_nontest['Plot'])
    tokens = vectorizer.get_feature_names_out()
    x_nontest_wordvec = pd.DataFrame(nontestIDF.toarray(), columns=tokens, index=x_nontest.index)
    x_nontest = pd.concat([x_nontest, x_nontest_wordvec], axis=1)
    x_nontest.drop(columns=['Plot'], inplace=True)
    
    #Made with the help of Deepseek
    # Function to generate binary columns from a list column
    def binarize_column(df, column_name, possible_values):
        return pd.DataFrame({
            value: df[column_name].apply(lambda x: 1 if value in x else 0)
            for value in possible_values
        })
    
    to_be_encoded = ['Cast', 'Directors', 'Distributors', 'Producers', 'Genre']
    
    for feat in to_be_encoded:
        vectorizer_b = joblib.load(f'{feat}_vectorizer.joblib')
        tokens_b = vectorizer_b.get_feature_names_out()
        
        nontestIDF = vectorizer_b.transform(x_nontest[feat])
        x_nontest_wordvec = pd.DataFrame(nontestIDF.toarray(), columns=tokens_b, index=x_nontest.index)
        x_nontest = pd.concat([x_nontest, x_nontest_wordvec], axis=1)

    x_nontest = x_nontest.drop(columns=to_be_encoded)

    #MPAA ratings
    Ratings = {
            'G': 1,
            'PG': 2,
            'PG-13': 3,
            'R': 4,
            'NC-17': 5,
            'No Certificate': 6}
    
    x_nontest['MPA Rating'] = x_nontest['MPA Rating'].map(Ratings)
    
    #PCA
    x_nontest_pca = pca.transform(x_nontest[tokens])[:, :27]
    pca_columns = [f'PC{i+1}' for i in range(x_nontest_pca.shape[1])]
    
    x_nontest = pd.concat([
        x_nontest.drop(columns=tokens),
        pd.DataFrame(x_nontest_pca, columns=pca_columns, index=x_nontest.index)
    ], axis=1)
    
    #Budget
    Budgets = {
            'Very low': 1,
            'Low': 2,
            'Mid': 3,
            'High': 4}
    x_nontest['Budget'] = x_nontest['Budget'].map(Budgets)
    
    #scale minutes
    x_nontest['Runtime in Minutes'] = np.log(x_nontest['Runtime in Minutes'])
    Scaler.transform(x_nontest[['Runtime in Minutes']])
    
    x_nontest = x_nontest.drop(columns=['Title', 'Year'])
    
    prediction  = None
    
    prediction = loaded_model.predict(x_nontest)[0]
    confidence =np.max(loaded_model.predict_proba(x_nontest), axis=1) 

    
    title = df['Title'].iloc[0]
    runtime = df['Runtime in Minutes'].iloc[0]
    genre = df['Genre'].iloc[0]
    directors = df['Directors'].iloc[0]
    distributors = df['Distributors'].iloc[0]
    producers = df['Producers'].iloc[0]
    mpa_rating = df['MPA Rating'].iloc[0]
    rating = prediction
    plot = df['Plot'].iloc[0]
    cast = df['Cast'].iloc[0]
    budget = df['Budget'].iloc[0]
    imgs = ["https://upload.wikimedia.org/wikipedia/commons/c/c4/Phillips%281804%29_p653_-_St_Giles%27s_Church_-_Sand_O%21.jpg",
            "https://upload.wikimedia.org/wikipedia/commons/1/17/Homegrown_green_tomato_03.jpg",
            "https://upload.wikimedia.org/wikipedia/commons/8/80/NASA_astronaut_Leland_D._Melvin_with_his_dogs_Jake_and_Scout.jpg",
            "https://images.pexels.com/photos/1762576/pexels-photo-1762576.jpeg?auto=compress&cs=tinysrgb&w=1260&h=750&dpr=2",
            "https://upload.wikimedia.org/wikipedia/commons/a/a2/Einstein_oppenheimer.jpg",
            "https://images.pexels.com/photos/753697/pexels-photo-753697.jpeg?auto=compress&cs=tinysrgb&w=1260&h=750&dpr=2"]
    
    
    html_content =  f"""
    <!DOCTYPE html>
    <html lang="en">
    <head>
        <meta charset="UTF-8">
        <title>{title} - Movie Details</title>
        <style>
            body {{
                font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
                background-color: #121212;
                color: #FFFFFF;
                margin: 0;
                padding: 0;
            }}
            .container {{
                max-width: 900px;
                margin: 40px auto;
                padding: 20px;
                background-color: #1e1e1e;
                border-radius: 10px;
                box-shadow: 0 4px 8px rgba(0, 0, 0, 0.3);
            }}
            header {{
                display: flex;
                align-items: center;
                gap: 20px;
                margin-bottom: 30px;
            }}
            .poster img {{
                width: 200px;
                height: auto;
                border-radius: 5px;
            }}
            .movie-info h1 {{
                margin: 0;
                font-size: 2.5em;
                color: #FFD700;
            }}
            .movie-info .meta {{
                margin-top: 10px;
                font-size: 1em;
                color: #BBBBBB;
            }}
            .section {{
                margin-top: 30px;
            }}
            .section h2 {{
                border-bottom: 2px solid #333333;
                padding-bottom: 10px;
                margin-bottom: 15px;
                color: #FFD700;
            }}
            .section p {{
                line-height: 1.6;
                color: #DDDDDD;
            }}
            .rating {{
                font-size: 1.5em;
                font-weight: bold;
                color: #00FF7F;
            }}
            @media (max-width: 600px) {{
                header {{
                    flex-direction: column;
                    align-items: flex-start;
                }}
                .poster img {{
                    width: 100%;
                }}
            }}
        </style>
    </head>
    <body>
        <div class="container">
            <header>
                <div class="poster">
                    <img src="{random.choice(imgs)}" alt="{title} Poster">
                </div>
                <div class="movie-info">
                    <h1>{title}</h1>
                    <div class="meta">
                        <p><strong>Runtime:</strong> {runtime} minutes</p>
                        <p><strong>MPA Rating:</strong> {mpa_rating}</p>
                        <p><strong>Budget:</strong> {budget}</p>
                        <p><strong>Genre:</strong> {genre}</p>
                    </div>
                </div>
            </header>
            <main>
                <section class="section">
                    <h2>Plot</h2>
                    <p>{plot}</p>
                </section>
                <section class="section">
                    <h2>Cast</h2>
                    <p>{cast}</p>
                </section>
                <section class="section">
                    <h2>Directors</h2>
                    <p>{directors}</p>
                </section>
                <section class="section">
                    <h2>Producers</h2>
                    <p>{producers}</p>
                </section>
                <section class="section">
                    <h2>Distributors</h2>
                    <p>{distributors}</p>
                </section>
                <section class="section">
                    <h2>Rating</h2>
                    <p class="rating">{rating} (with {confidence: .2%} confidence)</p>
                </section>
            </main>
        </div>
    </body>
    </html>
    """
    
    
    
    
    # JavaScript to open a new tab and write the HTML
    js_code = f"""
    var newWindow = window.open();
    newWindow.document.write(`{html_content}`);
    newWindow.document.close();
    """
    # Execute JS
    display(Javascript(js_code))
```

```python
import pandas as pd
from ipywidgets import widgets, Layout

common_style = {'description_width': 'initial'}

# Widgets
title = widgets.Text(
    placeholder='Type something here...',
    description='FILM TITLE:',
    style=common_style
)

plot = widgets.Textarea(
    placeholder='Enter Plot...',
    description='Plot:',
    style=common_style,
    layout=Layout(height='100px')
)

runtime = widgets.IntText(
    value=120,
    description='Runtime (in Minutes):',
    style=common_style
)

mpaa = widgets.RadioButtons(
    options=MPAA,
    value='G',
    description='MPAA Rating:',
    style=common_style
)

budget = widgets.RadioButtons(
    options=['Very low', 'Low', 'Mid', 'High'],
    value='Mid',
    description='Budget Level:',
    style=common_style
)

genres = widgets.SelectMultiple(
    options=GENRES,
    value=['adventure'],
    description='Genres:',
    style=common_style
)

cast = widgets.SelectMultiple(
    options=ACTORS,
    value=['adam sandler'],
    description='Actors:',
    style=common_style
)

director = widgets.Combobox(
    placeholder='Type or select an option',
    options=DIRECTORS,
    description='Director:',
    ensure_option=True,
    style=common_style
)

producer = widgets.Combobox(
    placeholder='Type or select an option',
    options=PRODUCERS,
    description='Production Company:',
    ensure_option=True,
    style=common_style
)

distributor = widgets.Combobox(
    placeholder='Type or select an option',
    options=DISTS,
    description='Distribution Company:',
    ensure_option=True,
    style=common_style
)

# Output area
output = widgets.Output()

# Submit button
button = widgets.Button(
    description='Submit',
    tooltip='Click to submit values',
    button_style='primary'
)

# Collect widget values into a one-row DataFrame without list brackets
def on_submit_clicked(b):
    with output:
        output.clear_output()
        global df
        data = {
            'Title': title.value,
            'Year':2024,
            'Plot': plot.value,
            'Budget': budget.value,
            'Cast': ",".join(cast.value),
            'MPA Rating': mpaa.value,
            'Directors': director.value,
            'Distributors': distributor.value,
            'Producers': producer.value,
            'Genre': ",".join(genres.value),
            'Runtime in Minutes': runtime.value}
        df = pd.DataFrame([data])
        everything()

button.on_click(on_submit_clicked)
widgets.VBox([title, plot, runtime, mpaa, budget, genres, cast, director, producer, distributor, button, output])
```
